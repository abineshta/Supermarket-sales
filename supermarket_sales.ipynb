## Insight 6

```python
#Splitting the date in date, month and year
df['Date'] = pd.to_datetime(df['Date'])
df['year'] = pd.DatetimeIndex(df['Date']).year
df['month'] = pd.DatetimeIndex(df['Date']).month
```

```python
df_agg = df.groupby(['month', 'Branch']).sum().reset_index()
df_pivoted = df_agg.pivot('month', 'Branch', 'Total')

sns.heatmap(df_pivoted).set_title("Monthly Wise Sales Revenue Over Branches",size = 17)
```

### Through this heat map, we can identify the variation in the revenues over branches in different months. We can figure out that there is some kind of seasonal demand in the market as said in the above insight. For all three branches, the first month of the year 2019 has a good sales numbers.

###  However, the second month is creating a huge drop for the branches except for B, which has a drop with less rate. In the third month, we can see another significant rise in the branches except for B. We might expect branch B to rise in the following months. Suppose, if the rise does not take place significantly in this branch, we must find the problem there. Through this measure, we can ensure to remain on the safer side.

# Modeling

### Know about the data before you jump in...

```python
df.info()
```

```python
df.describe()
```

```python
da= df[["Total","Rating"]]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as py
```

### Checking the relationship between both variables

```python
sns.pairplot(df,x_vars="Total",y_vars="Rating", height = 10)
```

```python
x = da["Total"]
y= da["Rating"]
```

### Selecting the test and train data and making it in 2D array

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size = 0.7, random_state=100)
```

```python
x_train.shape
```

```python
x_train = x_train.values.reshape(-1,1)
```

```python
x_train.shape
```

```python
y_test = y_test.values.reshape(-1,1)
```

```python
x_test = x_test.values.reshape(-1,1)
```

```python
y_train = y_train.values.reshape(-1,1)
```

```python
y_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

## For training and its accuracy with predictions

```python
lr.fit(x_train,y_train)
```

```python
y_pred_tr = lr.predict(x_train)
```

```python
y_pred_tr
```

```python
from sklearn.metrics import r2_score, mean_squared_error
```

```python
# Calculate mean square error
msetr = mean_squared_error(y_train,y_pred_tr)
msetr
```

```python
# Calculate R square value
rsq_tr = r2_score(y_train,y_pred_tr)
rsq_tr
```

```python
print('mean squared error :',msetr)
print('r square :',rsq_tr)
```

```python
plt.scatter(x_train,y_train, color="Blue")
plt.plot(x_train,y_pred_tr, color = 'red')
plt.xlabel("Total")
plt.ylabel("Rating")
plt.title('Prediction with train data')
plt.show()
```

## For training and its accuracy with predictions

```python
lr.fit(x_test,y_test)
```

```python
lr.predict([[2.9]])
```

```python
y_pred_te = lr.predict(x_test)
```

```python
y_pred_te
```

```python
from sklearn.metrics import r2_score, mean_squared_error
```

```python
# Calculate mean square error
msete = mean_squared_error(y_test,y_pred_te)
msete
```

```python
# Calculate R square value
rsq_te = r2_score(y_test,y_pred_te)
rsq_te
```

```python
print('mean squared error :',msete)
print('r square :',rsq_te)
```

```python
plt.scatter(x_test,y_test, color="Blue")
plt.plot(x_test,y_pred_te, color = 'red')
plt.xlabel("Total")
plt.ylabel("Rating")
plt.title('Prediction with test data')
plt.show()
```

### Model for testing
